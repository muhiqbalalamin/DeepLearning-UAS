## Building an Input Pipeline

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# ─── 1. LOAD DAN PREPROCESS ─────────────────────────────────────────────
housing = fetch_california_housing()
X, y = housing.data, housing.target.reshape(-1, 1)

X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full, random_state=42)

scaler = StandardScaler().fit(X_train)
X_mean, X_std = scaler.mean_, scaler.scale_

# Gabung data dan simpan dalam CSV terbagi
def save_to_csv_parts(data, prefix, header, n_parts=10):
    os.makedirs("datasets/housing", exist_ok=True)
    path_fmt = f"datasets/housing/{prefix}_part_{{:02d}}.csv"
    filepaths = []
    for i, part in enumerate(np.array_split(data, n_parts)):
        filepath = path_fmt.format(i)
        np.savetxt(filepath, part, fmt="%.5f", delimiter=",", header=header, comments="")
        filepaths.append(filepath)
    return filepaths

header = ",".join(housing.feature_names + ["MedianHouseValue"])
train_files = save_to_csv_parts(np.c_[X_train, y_train], "train", header, 20)
valid_files = save_to_csv_parts(np.c_[X_valid, y_valid], "valid", header, 10)
test_files  = save_to_csv_parts(np.c_[X_test, y_test],   "test",  header, 10)

# ─── 2. PARSE DAN LOAD DATASET ──────────────────────────────────────────
n_inputs = X_train.shape[1]

@tf.function
def preprocess(line):
    defaults = [0.] * n_inputs + [tf.constant([], dtype=tf.float32)]
    fields = tf.io.decode_csv(line, record_defaults=defaults)
    x = tf.stack(fields[:-1])
    y = tf.stack(fields[-1:])
    return (x - X_mean) / X_std, y

def csv_reader_dataset(filepaths, repeat=1, batch_size=32,
                       n_readers=5, n_read_threads=None,
                       shuffle_buffer_size=10000, n_parse_threads=5):
    dataset = tf.data.Dataset.list_files(filepaths).repeat(repeat)
    dataset = dataset.interleave(
        lambda fp: tf.data.TextLineDataset(fp).skip(1),
        cycle_length=n_readers, num_parallel_calls=n_read_threads)
    dataset = dataset.shuffle(shuffle_buffer_size)
    dataset = dataset.map(preprocess, num_parallel_calls=n_parse_threads)
    return dataset.batch(batch_size).prefetch(1)

# ─── 3. TRAINING DAN EVALUASI MODEL ─────────────────────────────────────
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)

train_set = csv_reader_dataset(train_files, batch_size=32)
valid_set = csv_reader_dataset(valid_files)
test_set  = csv_reader_dataset(test_files)

model = keras.models.Sequential([
    keras.layers.Dense(30, activation="relu", input_shape=[n_inputs]),
    keras.layers.Dense(1),
])

model.compile(loss="mse", optimizer=keras.optimizers.SGD(learning_rate=1e-3))
model.fit(train_set, steps_per_epoch=len(X_train) // 32, epochs=10, validation_data=valid_set)
model.evaluate(test_set, steps=len(X_test) // 32)

# ─── 4. PREDIKSI ────────────────────────────────────────────────────────
predictions = model.predict(test_set.map(lambda X, y: X), steps=len(X_test) // 32)

# ─── 5. CUSTOM TRAINING LOOP DENGAN TF.FUNCTION ────────────────────────
optimizer = keras.optimizers.Nadam(learning_rate=0.01)
loss_fn = keras.losses.MeanSquaredError()  # FIXED: now using loss instance

@tf.function
def train_loop(model, n_epochs, batch_size=32):
    train_set = csv_reader_dataset(train_files, repeat=n_epochs, batch_size=batch_size)
    n_steps = (len(X_train) // batch_size) * n_epochs
    step = 0
    for X_batch, y_batch in train_set.take(n_steps):
        step += 1
        with tf.GradientTape() as tape:
            y_pred = model(X_batch)
            main_loss = loss_fn(y_batch, y_pred)
            total_loss = tf.add_n([main_loss] + model.losses)
        grads = tape.gradient(total_loss, model.trainable_variables)
        optimizer.apply_gradients(zip(grads, model.trainable_variables))
        if step % 100 == 0:
            tf.print("Step", step, "/", n_steps)

train_loop(model, n_epochs=5)

# ─── 6. OPSIONAL: CETAK API tf.data.Dataset ────────────────────────────
print("\n● Fungsi API tf.data.Dataset:")
for m in dir(tf.data.Dataset):
    if not (m.startswith("_") or m.endswith("_")):
        func = getattr(tf.data.Dataset, m)
        if hasattr(func, "__doc__"):
            print(f"● {m}() - {func.__doc__.splitlines()[0]}")
```

## TFRecord

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_images

# ─── 1. Tulis & Baca TFRecord Sederhana ───────────────────────────────
with tf.io.TFRecordWriter("my_data.tfrecord") as f:
    f.write(b"This is the first record")
    f.write(b"And this is the second record")

# Baca record
for item in tf.data.TFRecordDataset(["my_data.tfrecord"]):
    print(item.numpy())

# ─── 2. TFRecord Paralel dari Beberapa File ────────────────────────────
filepaths = [f"my_test_{i}.tfrecord" for i in range(5)]
for i, path in enumerate(filepaths):
    with tf.io.TFRecordWriter(path) as f:
        for j in range(3):
            f.write(f"File {i} record {j}".encode("utf-8"))

# Baca paralel (multi-thread)
dataset = tf.data.TFRecordDataset(filepaths, num_parallel_reads=3)
for item in dataset:
    print(item.numpy())

# ─── 3. TFRecord dengan Kompresi GZIP ──────────────────────────────────
options = tf.io.TFRecordOptions(compression_type="GZIP")
with tf.io.TFRecordWriter("my_compressed.tfrecord", options) as f:
    f.write(b"This is the first record")
    f.write(b"And this is the second record")

for item in tf.data.TFRecordDataset(["my_compressed.tfrecord"], compression_type="GZIP"):
    print(item.numpy())

# ─── 4. Encode dan Decode Gambar dalam TFRecord ────────────────────────
img = load_sample_images()["images"][0]

# Tampilkan gambar asli
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()

# Encode gambar ke JPEG dan simpan sebagai TFRecord Example
data = tf.io.encode_jpeg(img)
feature = tf.train.Feature(bytes_list=tf.train.BytesList(value=[data.numpy()]))
example = tf.train.Example(features=tf.train.Features(feature={"image": feature}))
serialized = example.SerializeToString()

# Parse dan decode dari serialized example
desc = {"image": tf.io.VarLenFeature(tf.string)}
parsed = tf.io.parse_single_example(serialized, desc)
decoded_img = tf.io.decode_jpeg(parsed["image"].values[0])

# Tampilkan hasil decode
plt.imshow(decoded_img)
plt.axis("off")
plt.title("Decoded Image")
plt.show()

# ─── 5. Serialize dan Parse Tensor ─────────────────────────────────────
tensor = tf.constant([[0., 1.], [2., 3.], [4., 5.]])
s = tf.io.serialize_tensor(tensor)
print("Serialized Tensor:", s.numpy())

parsed_tensor = tf.io.parse_tensor(s, out_type=tf.float32)
print("Parsed Tensor:\n", parsed_tensor.numpy())
```

## TF Transform dan TensorFlow Datasets

```python
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
import matplotlib.pyplot as plt

# ─── 1. Load MNIST dan Visualisasi Sample ───────────────────────────────
datasets = tfds.load(name="mnist", as_supervised=True)
mnist_train, mnist_test = datasets["train"], datasets["test"]

# Ambil 5 gambar pertama dari dataset untuk ditampilkan
plt.figure(figsize=(6, 3))
for index, (img, label) in enumerate(mnist_train.take(5)):
    plt.subplot(1, 5, index + 1)  # Fix: gunakan index, bukan label
    plt.imshow(img.numpy().squeeze(), cmap="binary")
    plt.title(int(label.numpy()))
    plt.axis("off")
plt.show()

# ─── 2. Persiapkan Dataset ──────────────────────────────────────────────
BATCH_SIZE = 32
mnist_train = mnist_train.map(lambda x, y: (tf.cast(x, tf.float32) / 255.0, y))
mnist_train = mnist_train.shuffle(1000).batch(BATCH_SIZE).prefetch(1)

mnist_test = mnist_test.map(lambda x, y: (tf.cast(x, tf.float32) / 255.0, y))
mnist_test = mnist_test.batch(BATCH_SIZE).prefetch(1)

# ─── 3. Build & Compile Model ───────────────────────────────────────────
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28, 1]),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])

model.compile(loss="sparse_categorical_crossentropy",
              optimizer=keras.optimizers.Adam(1e-3),
              metrics=["accuracy"])

# ─── 4. Train Model ─────────────────────────────────────────────────────
history = model.fit(mnist_train,
                    validation_data=mnist_test,
                    epochs=5,
                    steps_per_epoch=60000 // BATCH_SIZE,
                    validation_steps=10000 // BATCH_SIZE)

# ─── 5. Evaluasi Akhir dan Plot ─────────────────────────────────────────
results = model.evaluate(mnist_test, steps=10000 // BATCH_SIZE, return_dict=True)
print(f"\n✅ Accuracy: {results['accuracy']:.4f} | "
      f"Loss: {results['loss']:.4f}")

# Plot akurasi dan loss
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(history.history["accuracy"], label="Train Acc")
plt.plot(history.history["val_accuracy"], label="Val Acc")
plt.title("Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.title("Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
plt.show()
```

# **Proses Pelatihan Model MNIST dengan TensorFlow dan TFDS**

## 1. Mengimpor Dataset MNIST
Dataset MNIST dimuat langsung menggunakan tensorflow_datasets (tfds) dan dibagi menjadi dua bagian:

* Train set untuk pelatihan

* Test set untuk evaluasi akhir

## 2. Visualisasi Data
Lima sampel gambar digit tulisan tangan ditampilkan untuk memberikan gambaran awal terhadap bentuk data yang akan digunakan.

## 3. Prapemrosesan Data
* Gambar dinormalisasi ke rentang [0.0,1.0] menggunakan:

tf.cast(x, tf.float32) / 255.0

* Dataset dilatih menggunakan teknik:
    * Batching (batch(32))

    * Shuffling (mengacak urutan data)

    * Prefetching (untuk efisiensi pipeline)

## 4. Membangun Model
Model dikonstruksi menggunakan Sequential API dengan arsitektur sederhana:

* Flatten layer: mengubah input gambar 28x28 menjadi vektor 1 dimensi

* Dense(128, activation='relu'): layer tersembunyi

* Dense(10, activation='softmax'): layer output untuk klasifikasi 10 kelas (digit 0–9)

## 5. Kompilasi dan Pelatihan
Fungsi loss: sparse_categorical_crossentropy

* Optimizer: Adam dengan learning rate 0.001

* Metrik evaluasi: akurasi (accuracy)

* Model dilatih selama 5 epoch

## 6. Evaluasi dan Visualisasi
* Evaluasi dilakukan terhadap validation set dan test set

* Plot akurasi dan loss per epoch ditampilkan untuk memantau performa model sepanjang pelatihan
